```python
import numpy as np
import re
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
geneName = "PRDM2"

res_rand_tfs_log = "/Users/Dinesh/Dropbox/Github/predicting_gex_with_nn_v2/Output/new/{}_t0.5/{}_wRandomTFs.log".format(geneName.lower(), geneName)
res_orig_tfs_log = "/Users/Dinesh/Dropbox/Github/predicting_gex_with_nn_v2/Output/new/{}_t0.5/{}.log".format(geneName.lower(), geneName)
```

```python
'''Example of a line in the file:
x = "2017-10-23 21:42:50,314: __main__    : INFO    :     gene:SKI, mode:tfs, lr:0.500, train_pcc:0.763, train_loss:0.127, test_pcc:0.778, test_loss:0.186"'''
def get_res_dict(alogfile, mode):
    handleIn = open(alogfile)
    lines = handleIn.readlines()
    handleIn.close()
    
    res_dict = {"train_pcc":[], "train_loss":[], "test_pcc":[], "test_loss":[]}
    for aline in lines:
        if ("gene:{}, mode:{}, ".format(geneName, mode) in aline):  # IMPORTANT: comma at the end to separate dhss from dhss_tfs mode
            x = aline[aline.index("gene"):]  # returns "gene:..."
            x = re.split(", ", x)
            d = dict([re.split(":", i) for i in x])  # returns {'gene': 'SKI', 'lr': '0.500', 'mode': 'tfs', ..}
            res_dict["train_pcc"].append(float(d["train_pcc"]))  
            res_dict["test_pcc"].append(float(d["test_pcc"]))
            res_dict["train_loss"].append(float(d["train_loss"]))
            res_dict["test_loss"].append(float(d["test_loss"]))
            
    return res_dict

def select_best_perf_over_lrs(res_dict, len_lrs, by="test_loss"):  # perf == performances; lrs = learning rates
    # len_lrs is the length of the lr vector i.e. number of lrs we have log info for.
    assert by in res_dict.keys() + ["test_pcc_over_loss", "train_pcc_over_loss", "train_and_test_loss_over_pcc"]
    best_perfs = []
    if (by in res_dict.keys()):
        d = res_dict[by]
    elif (by in ["test_pcc_over_loss"]):
        d1 = np.array(res_dict["test_pcc"])
        d2 = np.array(res_dict["test_loss"])
        d = d1/d2
    elif (by in ["train_pcc_over_loss"]):
        d1 = np.array(res_dict["train_pcc"])
        d2 = np.array(res_dict["train_loss"])
        d = d1/d2
    elif (by == "train_and_test_loss_over_pcc"):
        c_train = np.array(res_dict["train_loss"])  # c == cost
        c_test = np.array(res_dict["test_loss"])
        rho_train = np.array(res_dict["train_pcc"])
        rho_test = np.array(res_dict["test_pcc"])
        d = ((c_train/rho_train)*(1./42)) + ((c_test/rho_test)*(1./12))  # the denominators represent number of train and test examples
    else:
        raise Exception()
    assert len(d) % len_lrs == 0
    print(">>{}".format(len(d)/len_lrs))
    for i in range(0, len(d)/len_lrs): 
        j = i*len_lrs
        tup_perfs = np.array(d[j:j+len_lrs]) # np.array([d[j], d[j+1], d[j+2]])
        if (by in ["test_loss", "train_loss", "train_and_test_loss_over_pcc"]):
            best_perfs.append( min(tup_perfs) )
        elif (by in ["test_pcc", "train_pcc", "test_pcc_over_loss", "train_pcc_over_loss"]):
            best_perfs.append( max(tup_perfs) )
        else:
            raise Exception("undefined..")
    return best_perfs            

def get_best_perfs(chose_by, len_lrs, mode):
    res_dict_rand = get_res_dict(alogfile=res_rand_tfs_log, mode=mode)
    best_perfs_rand = select_best_perf_over_lrs(res_dict_rand, len_lrs, by=chose_by)

    res_dict_orig = get_res_dict(alogfile=res_orig_tfs_log, mode=mode)
    best_perfs_orig = select_best_perf_over_lrs(res_dict_orig, len_lrs, by=chose_by)
    return best_perfs_rand, best_perfs_orig
```

```python
def get_corrVals(is_with_random_tfs):
    if not (is_with_random_tfs):
        afile = res_orig_tfs_log
        searchfor = "    TF selected"  # mind the gaps
    else:
        afile = res_rand_tfs_log
        searchfor = "random TF selected"
    
    handleIn = open(afile)
    lines = handleIn.readlines()
    handleIn.close()
    
    corrs = []
    for aline in lines:
        if (searchfor in aline):
            x = aline[aline.index(searchfor):]  # return for eg: "random TF selected 0, ZNF143:0.36"
            x = re.split(", ", x)[1]
            corrs.append(float(re.split(":", x)[1]))
    return corrs
```

```python
mode = "dhss_tfs"#_tfs"#_tfs" # can be one of dhss, tfs or dhss_tfs
len_lrs = 6

fig = plt.figure(0, (10, 15))
sns.set(font_scale=1.2)
plt.suptitle("Model performances for {} (mode: {})".format(geneName, mode))

for plot_num, chose_by in enumerate(["train_pcc", "test_pcc", 
                                     "train_loss", "test_loss", 
                                     "train_pcc_over_loss", "test_pcc_over_loss", 
                                     "train_and_test_loss_over_pcc"]):
    if (plot_num == 7): # only plotting 7 subplots
        break
    plt.subplot(4,2,plot_num+1)
    best_perfs_rand, best_perfs_orig = get_best_perfs(chose_by, len_lrs, mode, bw=0.001)
    sns.kdeplot(np.array(best_perfs_rand), label="random TFs")
    if (plot_num <=3) or (plot_num == 6):
        plt.plot([best_perfs_orig[0], best_perfs_orig[0]], [0, 10], label="orig TFs")  # (x1, x2), (y1, y2)
    else:
        plt.plot([best_perfs_orig[0], best_perfs_orig[0]], [0, 1], label="orig TFs")  # (x1, x2), (y1, y2)
    plt.xlabel("{}".format(chose_by))
    plt.ylabel("Density")
    plt.legend()
```

```python
'''Plot the correlation values'''
rand_corrs = get_corrVals(is_with_random_tfs=True)
orig_corrs = get_corrVals(is_with_random_tfs=False)

sns.kdeplot(np.array(rand_corrs), label="random tfs")
sns.kdeplot(np.array(orig_corrs), label="orig tfs")
plt.legend()
plt.xlabel("Correlation of TFs with {} gex".format(geneName))
plt.ylabel("Density")
plt.title("Correlation Density Plots")
```
